```python
import pandas as pd
import numpy as np

def generate_dummy_dataframe():
    """
    Generates a dummy pandas DataFrame with specified columns and random values.

    Returns:
        pd.DataFrame: A dummy DataFrame.
    """
    # Dates from Jan 1, 2026 to Mar 31, 2026
    dates = pd.date_range(start='2026-01-01', end='2026-03-31')

    # Business Units
    bus = ['CPS', 'SBS']

    # Tiers (6-8 random tiers)
    tiers = [f'Tier {chr(65 + i)}' for i in range(7)] # A, B, C, D, E, F, G

    # Sites (5-7 random locations)
    sites = [f'Location {i+1}' for i in range(6)] # 1, 2, 3, 4, 5, 6

    # Create a list to hold all data rows
    data = []

    for date in dates:
        for bu in bus:
            for tier in tiers:
                for site in sites:
                    calls_handled = np.random.randint(50, 500) # Random calls handled
                    data.append({
                        'date': date,
                        'bu': bu,
                        'tier': tier,
                        'site': site,
                        'calls_handled': calls_handled
                    })

    df = pd.DataFrame(data)
    return df
```

```python
import random

def generate_additional_dataframe(base_df):
    """
    Generates an additional pandas DataFrame with extra columns related to offers and calls.

    Args:
        base_df (pd.DataFrame): The base DataFrame containing 'date', 'bu', 'tier', 'site'.

    Returns:
        pd.DataFrame: A new DataFrame with added offer and calls data.
    """
    # Get unique combinations of date, bu, tier, site from the base_df
    unique_combinations = base_df[['date', 'bu', 'tier', 'site']].drop_duplicates()

    new_data = []
    possible_offers = [f'Offer {chr(65 + i)}' for i in range(10)] # e.g., Offer A to Offer J

    for _, row in unique_combinations.iterrows():
        date = row['date']
        bu = row['bu']
        tier = row['tier']
        site = row['site']

        # offers_present: 1-7 random offers
        num_offers_present = random.randint(1, 7)
        offers_present_list = random.sample(possible_offers, num_offers_present)
        offers_present_str = '|'.join(sorted(offers_present_list))

        # offers_extended: subset of offers_present
        num_offers_extended = random.randint(0, len(offers_present_list))
        offers_extended_list = random.sample(offers_present_list, num_offers_extended)
        offers_extended_str = '|'.join(sorted(offers_extended_list))

        # offers_accepted: subset of offers_extended
        num_offers_accepted = random.randint(0, len(offers_extended_list))
        offers_accepted_list = random.sample(offers_extended_list, num_offers_accepted)
        offers_accepted_str = '|'.join(sorted(offers_accepted_list))

        # calls: random number
        calls = np.random.randint(10, 1000)

        new_data.append({
            'date': date,
            'bu': bu,
            'tier': tier,
            'site': site,
            'offers_present': offers_present_str,
            'offers_extended': offers_extended_str,
            'offers_accepted': offers_accepted_str,
            'calls': calls
        })

    new_df = pd.DataFrame(new_data)
    return new_df

# Generate the new DataFrame using elig_dat as the base
additional_df = generate_additional_dataframe(elig_dat)

# Display the first 5 rows and info of the new DataFrame
display(additional_df.head())
display(additional_df.info())
```

## Generate Detailed Offer Data

### Subtask:
Create a function to generate detailed offer data. For each unique (date, bu, tier, site) from elig_dat, this function will generate multiple offer rows (offers_present, offers_extended, offers_accepted) and distribute a total calls count among them, ensuring the sum of these calls is less than or equal to the calls_handled from elig_dat for that combination. Each offer will have a distinct set of offers presented, extended, and accepted.


**Reasoning**:
The subtask requires defining a function `generate_detailed_offer_data` to create granular offer data from `elig_dat`, distributing `calls_handled` across multiple new offer rows while ensuring the sum of calls for each group does not exceed the original `calls_handled`. I will implement the function as specified in the instructions, including handling the distribution of calls among the generated offer rows.



```python
import pandas as pd
import numpy as np
import random

def generate_detailed_offer_data(base_df, max_offers_per_group=5):
    """
    Generates detailed offer data by breaking down 'calls_handled' into multiple offer rows
    for each unique combination of 'date', 'bu', 'tier', and 'site'.

    Args:
        base_df (pd.DataFrame): The base DataFrame containing 'date', 'bu', 'tier', 'site', and 'calls_handled'.
        max_offers_per_group (int): The maximum number of offer rows to generate for each unique group.

    Returns:
        pd.DataFrame: A new DataFrame with detailed offer data.
    """
    detailed_data = []

    # Get unique combinations of date, bu, tier, site and their calls_handled
    unique_combinations = base_df[['date', 'bu', 'tier', 'site', 'calls_handled']].drop_duplicates()

    possible_offers = [f'Offer {chr(65 + i)}' for i in range(10)] # e.g., Offer A to Offer J

    for _, row in unique_combinations.iterrows():
        date = row['date']
        bu = row['bu']
        tier = row['tier']
        site = row['site']
        group_calls_handled = row['calls_handled']

        # Generate a random number of offer rows for this group
        num_offer_rows = random.randint(1, max_offers_per_group)

        call_values = [0] * num_offer_rows # Initialize with zeros

        if group_calls_handled > 0 and num_offer_rows > 0:
            if group_calls_handled >= num_offer_rows:
                # Use cut-point method to distribute calls, ensuring sum == group_calls_handled and each value >= 1.
                if num_offer_rows > 1:
                    # Generate num_offer_rows - 1 random cut points between 1 and group_calls_handled - 1
                    # Ensure enough unique points can be chosen
                    if group_calls_handled - 1 >= num_offer_rows - 1:
                        cut_points = sorted(random.sample(range(1, group_calls_handled), num_offer_rows - 1))
                        all_points = [0] + cut_points + [group_calls_handled]
                        call_values = [all_points[i+1] - all_points[i] for i in range(num_offer_rows)]
                    else:
                        # Fallback if not enough distinct cut points can be generated
                        call_values = [group_calls_handled // num_offer_rows + (1 if i < group_calls_handled % num_offer_rows else 0) for i in range(num_offer_rows)]
                        # Ensure minimum 1 call if possible
                        call_values = [max(1, c) for c in call_values]

                else: # num_offer_rows == 1
                    call_values = [group_calls_handled]
            else:
                # group_calls_handled < num_offer_rows. Distribute 1 call to 'group_calls_handled' rows
                # and 0 to the rest. Sum will be exactly group_calls_handled.
                assigned_indices = random.sample(range(num_offer_rows), group_calls_handled)
                for idx in assigned_indices:
                    call_values[idx] = 1
        # If group_calls_handled is 0, call_values remains all zeros, which is correct.

        # Shuffle the call_values to randomize their assignment to offer rows
        random.shuffle(call_values)

        for i in range(num_offer_rows):
            # b. Randomly select 1 to 7 offers for offers_present
            num_offers_present = random.randint(1, 7)
            offers_present_list = random.sample(possible_offers, num_offers_present)
            offers_present_str = '|'.join(sorted(offers_present_list))

            # c. Randomly select a subset of offers_present for offers_extended
            num_offers_extended = random.randint(0, len(offers_present_list))
            offers_extended_list = random.sample(offers_present_list, num_offers_extended)
            offers_extended_str = '|'.join(sorted(offers_extended_list))

            # d. Randomly select a subset of offers_extended for offers_accepted
            num_offers_accepted = random.randint(0, len(offers_extended_list))
            offers_accepted_list = random.sample(offers_extended_list, num_offers_accepted)
            offers_accepted_str = '|'.join(sorted(offers_accepted_list))

            # e. Assign calls value
            current_calls = call_values[i]

            # f. Append a dictionary
            detailed_data.append({
                'date': date,
                'bu': bu,
                'tier': tier,
                'site': site,
                'offers_present': offers_present_str,
                'offers_extended': offers_extended_str,
                'offers_accepted': offers_accepted_str,
                'calls': current_calls
            })

    new_df = pd.DataFrame(detailed_data)
    return new_df
```

**Reasoning**:
Now that the `generate_detailed_offer_data` function is defined, I will call it with `elig_dat` to create the new detailed offer DataFrame and then display its head and info to inspect the generated data as specified in the subtask.



```python
detailed_offer_df = generate_detailed_offer_data(elig_dat)
display(detailed_offer_df.head())
display(detailed_offer_df.info())
```

**Reasoning**:
I need to merge the `elig_dat` DataFrame (which contains `calls_handled`) with the newly created `detailed_offer_df` on the common key columns (`date`, `bu`, `tier`, `site`). This will combine the granular offer data with the original `calls_handled` for each group.



```python
combined_df = pd.merge(elig_dat, detailed_offer_df, on=['date', 'bu', 'tier', 'site'], how='left')
display(combined_df.head())
display(combined_df.info())
```

**Reasoning**:
The final part of the subtask is to verify that for each unique `(date, bu, tier, site)` group, the sum of `calls` from the offer rows does not exceed the `calls_handled` value for that group. I will achieve this by grouping the `combined_df` by the key columns, summing the `calls`, and then comparing it against the `calls_handled` from the original `elig_dat`.



```python
grouped_calls = combined_df.groupby(['date', 'bu', 'tier', 'site'])['calls'].sum().reset_index()
merged_verification = pd.merge(elig_dat, grouped_calls, on=['date', 'bu', 'tier', 'site'], how='inner')
merged_verification.rename(columns={'calls': 'sum_of_detailed_calls'}, inplace=True)

# Check if sum_of_detailed_calls <= calls_handled for all groups
verification_result = (merged_verification['sum_of_detailed_calls'] <= merged_verification['calls_handled']).all()

print(f"Verification Result: Is the sum of 'calls' for each group less than or equal to 'calls_handled'? {verification_result}")

# Optionally, display groups where the condition might be violated (should be empty if verification_result is True)
if not verification_result:
    display(merged_verification[merged_verification['sum_of_detailed_calls'] > merged_verification['calls_handled']])
```

```python
import pandas as pd
import numpy as np

def generate_dummy_dataframe():
    """
    Generates a dummy pandas DataFrame with specified columns and random values.

    Returns:
        pd.DataFrame: A dummy DataFrame.
    """
    # Dates from Jan 1, 2026 to Apr 30, 2026
    dates = pd.date_range(start='2026-01-01', end='2026-04-30')

    # Business Units
    bus = ['CPS', 'SBS']

    # Tiers (6-8 random tiers)
    tiers = [f'Tier {chr(65 + i)}' for i in range(7)] # A, B, C, D, E, F, G

    # Sites (5-7 random locations)
    sites = [f'Location {i+1}' for i in range(6)] # 1, 2, 3, 4, 5, 6

    # Create a list to hold all data rows
    data = []

    for date in dates:
        for bu in bus:
            for tier in tiers:
                for site in sites:
                    calls_handled = np.random.randint(50, 500) # Random calls handled
                    data.append({
                        'date': date,
                        'bu': bu,
                        'tier': tier,
                        'site': site,
                        'calls_handled': calls_handled
                    })

    df = pd.DataFrame(data)
    return df
```

**Reasoning**:

**Reasoning**:
To aggregate `combined_df` weekly, I first need to ensure the 'date' column is in datetime format. Then, I will create a new column representing the start of each week (Monday). Finally, I'll group the DataFrame by this new weekly column, sum 'calls_handled' and 'calls', and filter for the last four weeks.



```python
combined_df['date'] = pd.to_datetime(combined_df['date'])
combined_df['week_start'] = combined_df['date'].apply(lambda x: x - pd.Timedelta(days=x.weekday()))

weekly_aggregated_df = combined_df.groupby('week_start').agg({
    'calls_handled': 'sum',
    'calls': 'sum'
}).reset_index()

# Sort by week_start to easily get the last 4 weeks
weekly_aggregated_df = weekly_aggregated_df.sort_values('week_start', ascending=False)

# Filter for the last 4 weeks
last_4_weeks_df = weekly_aggregated_df.head(4).sort_values('week_start', ascending=True)

display(last_4_weeks_df.head())
```

**Reasoning**:
To aggregate `combined_df` monthly, I will first extract the month from the 'date' column, then group by this month and sum 'calls_handled' and 'calls' for each of the four months (January, February, March, April). Finally, I will create separate DataFrames for each month and display their heads.



```python
combined_df['month'] = combined_df['date'].dt.month

monthly_aggregated_df = combined_df.groupby('month').agg({
    'calls_handled': 'sum',
    'calls': 'sum'
}).reset_index()

# Convert the integer month to 'YYYY-MM' string format
# Assuming all data is for the year 2026 based on previous generation
monthly_aggregated_df['month_str'] = monthly_aggregated_df['month'].apply(lambda x: f'2026-{x:02d}')
monthly_aggregated_df.drop(columns=['month'], inplace=True) # Drop original month column
monthly_aggregated_df.rename(columns={'month_str': 'month'}, inplace=True) # Rename for consistency

# Create separate DataFrames for each month
jan_df = monthly_aggregated_df[monthly_aggregated_df['month'] == '2026-01']
feb_df = monthly_aggregated_df[monthly_aggregated_df['month'] == '2026-02']
mar_df = monthly_aggregated_df[monthly_aggregated_df['month'] == '2026-03']
apr_df = monthly_aggregated_df[monthly_aggregated_df['month'] == '2026-04']

print("January Aggregated Data:")
display(jan_df.head())

print("February Aggregated Data:")
display(feb_df.head())

print("March Aggregated Data:")
display(mar_df.head())

print("April Aggregated Data:")
display(apr_df.head())
```
